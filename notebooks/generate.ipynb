```python
import json
import os
import json
import torch
from diffusers import BriaFiboPipeline
from diffusers.modular_pipelines import ModularPipeline
```

```python
torch.set_grad_enabled(False)
```

```python
pipe = BriaFiboPipeline.from_pretrained("/home/ckenny/models/FIBO", torch_dtype=torch.bfloat16)
```

```python
pipe.to("cuda")
```

```python

prompt = json.load(open("/home/ckenny/models/FIBO/examples/outputs/generate.json", "r"))
```

prompt['background_setting'] = "dark park at night"
prompt['aesthetics']['color_scheme'] = "dark, desaturated, high contrast"
prompt['aesthetics']['mood_atmosphere'] = "spooky, mysterious, dramatic"
prompt['lighting']['conditions'] = "night time environment, with a bright point source light on the ground in front of the dog but out of frame providing most of the light"
prompt['lighting']['direction'] = "from below up onto the dog, providing so called campfire lighting, think of a person sitting around telling ghost stories with a flashlight under their chin"
prompt['lighting']['shadows'] = "high contrast, highlighting the dog's features, think of a person sitting around telling ghost stories with a flashlight under their chin"


```python
results_generate = pipe(prompt=str(prompt),
                        num_inference_steps=5,
                        guidance_scale=5,
                        width=256,
                        height=256,
                        #do_patching=True,
                        #negative_prompt=negative_prompt
)
```

```python
results_generate.images[0].save("image_generate_ft_6.png")
```
